```python
def one_hot_encode_inplace(train_df: pd.DataFrame, test_df: pd.DataFrame, final_df: pd.DataFrame, columns: list):
    combined = pd.concat([train_df, test_df, final_df], axis=0)
    combined = pd.get_dummies(combined, columns=columns, drop_first=False)

    train_rows = train_df.shape[0]
    test_rows = test_df.shape[0]

    train_df.loc[:, combined.columns] = combined.iloc[:train_rows, :]
    test_df.loc[:, combined.columns] = combined.iloc[train_rows:train_rows + test_rows, :]
    final_df.loc[:, combined.columns] = combined.iloc[train_rows + test_rows:, :]

    train_df.drop(columns=columns, inplace=True, errors='ignore')
    test_df.drop(columns=columns, inplace=True, errors='ignore')
    final_df.drop(columns=columns, inplace=True, errors='ignore')
```
